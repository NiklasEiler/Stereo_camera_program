```python
import numpy as np
import pandas as pd
```

```python
csv_file_path = 'my_dataframe1.csv'


data = pd.read_csv(csv_file_path)
data= data.drop('3', axis=1)
data
```

```python
data=data.dropna()
data=data.values
data
```

```python
color=np.round(((data[:,2]-data[:,2].min())/(data[:,2].max()-data[:,2].min())*255)).astype(int)
color
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Importing 3D plotting tools

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with three columns as X, Y, and Z
ax.scatter(data[:,0], data[:,1], data[:,2])



plt.ion()

plt.show()

plt.show(block=True)
```

```python
# Generate random 3D data
np.random.seed(42)
n_points = 100
x = np.random.rand(n_points)
y = np.random.rand(n_points)
z = np.random.rand(n_points)

# Generate a random 1D color array
colors = np.random.rand(n_points)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x, y, z, c=colors, cmap='viridis')

# Add a color bar to show the mapping of values to colors
cbar = fig.colorbar(scatter, ax=ax, label='Color')

# Set labels for the axes
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
```
