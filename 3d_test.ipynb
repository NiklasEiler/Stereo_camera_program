```python
# Generate random 3D data
np.random.seed(42)
n_points = 100
x = np.random.rand(n_points)
y = np.random.rand(n_points)
z = np.random.rand(n_points)

# Generate a random 1D color array
colors = np.random.rand(n_points)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x, y, z, c=colors, cmap='viridis')

# Add a color bar to show the mapping of values to colors
cbar = fig.colorbar(scatter, ax=ax, label='Color')

# Set labels for the axes
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
```

```python
import numpy as np

# Create a NumPy array with some NaN values
arr = np.array([[1.0, 2.0, np.nan],
                [4.0, np.nan, 6.0],
                [7.0, 8.0, 9.0]])

# Create a boolean mask where True corresponds to NaN values
nan_mask = np.isnan(arr)

# Use the mask to filter out NaN values
arr_no_nan = arr[~nan_mask]

print("Original array:")
print(arr)
print("\nArray without NaN values:")
print(arr_no_nan)
```
